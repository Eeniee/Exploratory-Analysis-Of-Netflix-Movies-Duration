```python
netflix_movies =netflix_df[['title','country','genre','release_year','duration']]
netflix_movies
```

```python
short_movies=netflix_movies[netflix_movies['duration']<60]
short_movies
```

```python
color=[]

for lab, row in netflix_movies.iterrows() :
    if row['genre']=='Children':
        color.append('blue')
    elif row['genre']=='Documentaries':
        color.append('pink')
    elif row['genre']=='Stand-Up':
        color.append('red')
    elif row['genre']=='Other' :
        color.append('orange')
    else :
        color.append('green')
```

```python
#plots=plt.scatter(r(xaxis),c(yaxis))
fig=plt.scatter(netflix_df['release_year'],netflix_df['duration'],c=color,alpha=0.2)
plt.xlabel('Release_year')
plt.ylabel('Duration(min)')
plt.title('MOVIES DURATION BY YEAR')
plt.grid(True)

plt.show()
```

```python
netflix_avg = netflix_df.groupby('release_year')['duration'].mean()

netflix_avg.plot(x='year',y='duration',kind='line',color='red',title='AVERAGE MOVIE DURATION OVER THE YEARS')
plt.xlabel('Year')
plt.ylabel('Duration')
plt.grid(True)

plt.show()
```
